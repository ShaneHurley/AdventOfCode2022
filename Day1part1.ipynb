```python
input = open("day1.txt", "r")
```

```python
elves=input.read()
```

```python
elves=elves.split("\n\n")
```

```python
totalPerElf = []
for elf in elves:
    elf = elf.split("\n")
    elf = list(filter(lambda x: len(x) > 0, elf))
    elf = [int(i) for i in elf]
    totalPerElf.append(sum(elf))
totalPerElf.sort(reverse=True)
print(totalPerElf)
```

```python
print(totalPerElf[0])
print(sum(totalPerElf[0:3]))
```
